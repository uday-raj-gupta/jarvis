```python
import pyttsx3
import speech_recognition as sr
import os
import urllib.parse 
import webbrowser
import datetime
import pywhatkit

chrome_path = "C:/Program Files/Google/Chrome/Application/chrome.exe"
webbrowser.register('chrome', None, webbrowser.BackgroundBrowser(chrome_path))
speaker = pyttsx3.init()
mic = sr.Recognizer()

speaker.say("Welcome to Jarvis")
speaker.runAndWait()

with sr.Microphone() as source:
    print("Start Speaking......")
    mic.pause_threshold = 1
    audio = mic.listen(source)

try:
    text = mic.recognize_google(audio)
    text = text.lower()
    print("You said:", text)

    #  OPEN NOTEPAD
    if "open notepad" in text:
        print("opening notepad....")
        speaker.say("Opening Notepad")
        os.system("notepad")

    #  OPEN WHATSAPP
    elif "open whatsapp" in text:
        print("opening whatsapp....")
        speaker.say("Opening Whatsapp")
        os.system("start whatsapp://")
        
    if "open youtube" in text:
       print("Opening YouTube...")
       speaker.say("Opening YouTube")
       webbrowser.get(chrome_path + " %s").open("https://www.youtube.com")
    elif "play song" in text:
        speaker.say("Which song should I play?")
        speaker.runAndWait()

        with sr.Microphone() as song_source:
            print("Say the song name...")
            song_audio = mic.listen(song_source)
            song_name = mic.recognize_google(song_audio)
            song_name = song_name.lower()
            print("Playing:", song_name)
            speaker.say(f"Playing {song_name} on YouTube")
            speaker.runAndWait()
            pywhatkit.playonyt(song_name)

    elif "search for" in text:
       query = text.replace("search for", "")
       url = "https://www.google.com/search?q=" + urllib.parse.quote(query)
       print(f"Searching for {query}")
       speaker.say(f"Searching for {query}")
       webbrowser.get(chrome_path + " %s").open(url)
  

    elif "what is time now" in text:
      time = datetime.datetime.now().strftime("%H:%M")
      print("Time is", time)
      speaker.say(f"The time is {time}")



    speaker.runAndWait()

except Exception as e:
    print("Could not understand your voice. Try again...")
```
